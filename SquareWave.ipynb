```python
import numpy as np
import matplotlib.pyplot as plt

omega_not = (2*np.pi)*300*10**6 # fundamental frequency , f = 300 MHz, wavelength = 1 m ( in air )

points = np.arange(-np.pi/(300*10**6), np.pi/(300*10**6), 2*np.pi/(1000 * 300*10**6)) 
fundamental = np.sin(points*omega_not)

summation = np.empty(len(fundamental))
plt.figure(1)
plt.title("Square Wave Approximation with 1, 3, 5, 7, 9, and 11 Harmonics")
for i in range(1, 12, 2):
    plt.plot(points, np.sin(points*i*omega_not)/i, label=f"{i}th harmonic")
    summation+=(np.sin(points*i*omega_not)/i)
plt.legend()
plt.plot(points, summation)

summation = np.empty(len(fundamental))
plt.figure(2)
plt.title("Square Wave Approximation with 10x previous Harmonics")
for i in range(1, 112, 2):
    #plt.plot(points, np.sin(points*i*omega_not)/i, label=f"{i}th harmonic")
    summation+=(np.sin(points*i*omega_not)/i)
plt.plot(points, summation)


summation = np.empty(len(fundamental))
plt.figure(3)
plt.title("Square Wave Approximation with 10x previous Harmonics")
for i in range(1, 1112, 2):
    #plt.plot(points, np.sin(points*i*omega_not)/i, label=f"{i}th harmonic")
    summation+=(np.sin(points*i*omega_not)/i)
plt.plot(points, summation)
```
